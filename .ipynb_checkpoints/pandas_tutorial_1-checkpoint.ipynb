
<H1> DATA FILTERING </H1>


Let's find the IMDB rating of The Godfather movie.
If we know the order of the movie in the data frame, we can simply add the 2nd element of the IMBD_Rating column of the data frame (where the first index is 0).
We write 1 for , if you remember, the first element in lists starts from 0):

```python
df['IMDB_Rating'][1]

```

However, if we do not have a chance to see the dataset of the Godfather movie, that is, there is a large dataset somewhere, then we can find it with code:

```python
df.loc[df['Series_Title']=="The Godfather"]["IMDB_Rating"]
```

Let's explain how we did this. Firstly
With df.loc[df['Series_Title']=="The Godfather"] we find the line where the Godfather movie is actually located:

```python
df.loc[df['Series_Title']=="The Godfather"]
```

But what we want is just the IMDB_Rating value, that is, we do not want the entire row, we select the "IMDB_Rating" column from this row.

There is a variable that returns to us with df.loc. See, that variable is actually a dataframe:

```python
type(df.loc[df['Series_Title']=="The Godfather"])
```

Then we use square brackets to simply select a column of a dataframe and select:

```python
df.loc[df['Series_Title']=="The Godfather"]["IMDB_Rating"]
```

Finally, let's see the movies with IMDB_Rating value over 8 and No_of_Votes over 100000:

```python
df.loc[(df['IMDB_Rating']>=8) & (df['No_of_Votes']>=100000)]
```

Now let's see the movies with IMDB_Rating value over 8 and Gross value (i.e. box office revenue) over 30000000:

```python
df.loc[(df['IMDB_Rating']>=8) & (df['Gross']>=30000000)]
```

It gives an error because it is of #object type:
df.dtypes

```python
# convert column "Gross" of a DataFrame
# to_numeric This function will try to change non-numeric 
# objects (such as strings) into integers or floating point numbers as appropriate.
df['Gross'] = df['Gross'].str.replace(",", "")
df["Gross"] = pd.to_numeric(df["Gross"])
type(df["Gross"][0])
```

```python
df.loc[(df['IMDB_Rating']>=8.0) & (df['Gross']>=500000000)]
```


<h2> Creating Dataframe Manually</h2>

So far we have created and used Dataframes from a file. Now we will learn how we can create a dataframe ourselves, if needed, from scratch without using any files.

We will create a sample array using the random library.


```python
import random

randomlist1 = random.sample(range(15, 25), 2) 
randomlist2  = random.sample(range(15, 25), 2) 

randomlist1
```

```python
randomlist2
```

```python
randomlistoflists = [randomlist1, randomlist2]
randomlistoflists
```

```python
columns = ["Temperature day 1","Temperature day 2"] 
mydataframe = pd.DataFrame(randomlistoflists,index = ["Tokyo","Germany"],columns = columns) 
mydataframe
```
